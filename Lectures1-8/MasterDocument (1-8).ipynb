```python
# create empty lists
good_movies = [] 
ok_movies = []
bad_movies = []

cut_off_1 = 7.0 # minimum rating for "good movie"
cut_off_2 = 6.0 # minimum rating for "ok movie"


title_1 = "The Umbrellas of Cherbourg"
imdb_rating_1 = 7.8 

title_2 = "M3gan" 
imdb_rating_2 = 6.3 

title_3 = "Psycho" 
imdb_rating_3 = 8.5 

title_4 = "Melania"
imdb_rating_4 = 1.3

# Divide movies between the three lists
# print a warning if we encounter a bad movie

if imdb_rating_1 >= cut_off_1:
    good_movies = good_movies + [title_1] # add movie to good_movies
elif imdb_rating_1 >= cut_off_2:
    ok_movies = ok_movies + [title_1] # add movie to ok_movies
else:
    bad_movies = bad_movies + [title_1] # add movie to bad_movies
    print("Warning! Bad movie alert:", title_1) # ALSO: print warning message

if imdb_rating_2 >= cut_off_1:
    good_movies = good_movies + [title_2]  
elif imdb_rating_2 >= cut_off_2:
    ok_movies = ok_movies + [title_2]
else:
    bad_movies = bad_movies + [title_2]
    print("Warning! Bad movie alert:", title_2)


if imdb_rating_3 >= cut_off_1:
    good_movies = good_movies + [title_3] 
elif imdb_rating_3 >= cut_off_2:
    ok_movies = ok_movies + [title_3]
else:
    bad_movies = bad_movies + [title_3]
    print("Warning! Bad movie alert:", title_3)


if imdb_rating_4 >= cut_off_1:
    good_movies = good_movies + [title_4] 
elif imdb_rating_4 >= cut_off_2:
    ok_movies = ok_movies + [title_4]
else:
    bad_movies = bad_movies + [title_4]
    print("Warning! Bad movie alert:", title_4)


print(good_movies)
print(ok_movies)
print(bad_movies)
```

### WHILE Loops: Don't know how many times it will run

<font size = "4">

Consider the function

$$f(x) = \frac{1}{x}$$

- We can continue to run values until $0 < f(x) \leq \epsilon$

```python
eps = 0.025

x_list = []
y_list = []


x = 1
while 1/x >= eps: ### If this line is TRUE then loop begins; if FALSE it will skip to scatter line
    x_list.append(x)
    y_list.append(1/x)
    x = x + 1


plt.scatter(x = x_list, y = y_list)
plt.title("Graph of f(x) = 1/x")
plt.xlabel('x')
plt.ylabel("y")
plt.show()
```

### Success and Failure

```python
# The next example simulates a discrete-time Bernoulli process.
# For each time step, we will perform an experiment leading either 
# to "Success" or "Failure"

import random 

possible_outcomes = ["Success", "Failure"]

# experiment
outcome = random.choice(possible_outcomes)
print(outcome)
```

```python
# While Loop

success_count = 0
t = 0.0
final_time = 1.0
dt = 0.1

# while t != final_time:  # <--- WARNING: using this instead will cause infinite loop!

while abs(t - final_time) > 1e-13:
    # Experiment
    random_outcome = random.choice(possible_outcomes)

    # If it was a success, count it by increasing the success_count variable
    if random_outcome == "Success":
        success_count = success_count + 1

    # Increase time to the next value
    t = t + dt

print(success_count)
```
