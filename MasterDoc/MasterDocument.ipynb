### WHILE Loops: Don't know how many times it will run

<font size = "4">

Consider the function

$$f(x) = \frac{1}{x}$$

- We can continue to run values until $0 < f(x) \leq \epsilon$

````python
eps = 0.025

x_list = []
y_list = []


x = 1
while 1/x >= eps: ### If this line is TRUE then loop begins; if FALSE it will skip to scatter line
    x_list.append(x)
    y_list.append(1/x)
    x = x + 1


plt.scatter(x = x_list, y = y_list)
plt.title("Graph of f(x) = 1/x")
plt.xlabel('x')
plt.ylabel("y")
plt.show()
````

````python
eps = 0.025

x_list = [] 
y_list = []


max_x = 100000

for x in range(1, max_x+1):
    if (1/x) <= eps:
        break
    x_list.append(x)
    y_list.append(1/x)

plt.scatter(x = x_list, y=y_list)
plt.xlabel("x values")
plt.ylabel("1/x")
plt.show()
````

### Success and Failure

````python
# The next example simulates a discrete-time Bernoulli process.
# For each time step, we will perform an experiment leading either 
# to "Success" or "Failure"

import random 

possible_outcomes = ["Success", "Failure"]

# experiment
outcome = random.choice(possible_outcomes)
print(outcome)
````

````python
# While Loop

success_count = 0
t = 0.0
final_time = 1.0
dt = 0.1

# while t != final_time:  # <--- WARNING: using this instead will cause infinite loop!

while abs(t - final_time) > 1e-13:
    # Experiment
    random_outcome = random.choice(possible_outcomes)

    # If it was a success, count it by increasing the success_count variable
    if random_outcome == "Success":
        success_count = success_count + 1

    # Increase time to the next value
    t = t + dt

print(success_count)
````

````python
def quadratic_formula(a, b, c):
    #body
    discrim = b**2 - 4*a*c
    x_plus = -b + np.sqrt(discrim)
    x_plus = x_plus/(2*a)

    x_minus = -b - np.sqrt(discrim)
    x_minus = x_minus/(2*a)

    #return two values as output
    return x_plus, x_minus

# Run
output = quadratic_formula(5,15,10) #Output is tuple
print(output[0])
print(output[1])

#or...
pos, neg = quadratic_formula(5,15,10)
print(pos)
````

````python
# Create random list with missing data

import random
lst_length = 10
num_missing = 3

# Create numerical data
data = [random.randint(0, 100) for i in range(lst_length)]

# Randomly choose places where there will be missing data
missing_indices = list(range(lst_length))
random.shuffle(missing_indices)
missing_indices = missing_indices[:num_missing]

# Randomly choose between "N/A" and None values for the missing data
missing_vals = random.choices(["N/A", None], k = num_missing)

# Replace the random components of data with missing values
for idx, val in zip(missing_indices, missing_vals):
    data[idx] = val


print(data)
````

````python
# Complete the function

def avg_valid(input_list):
    valid_num = 0
    valid_denom = 0

    for value in input_list:
        if type(value) != float and type(value) != int:
            continue
        valid_num = valid_num + value
        valid_denom = valid_denom + 1
    
    valid_avg = valid_num / valid_denom
    return valid_avg

print(avg_valid(data))
````
