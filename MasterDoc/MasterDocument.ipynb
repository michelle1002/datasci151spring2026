### WHILE Loops: Don't know how many times it will run

<font size = "4">

Consider the function

$$f(x) = \frac{1}{x}$$

- We can continue to run values until $0 < f(x) \leq \epsilon$

````python
eps = 0.025

x_list = []
y_list = []


x = 1
while 1/x >= eps: ### If this line is TRUE then loop begins; if FALSE it will skip to scatter line
    x_list.append(x)
    y_list.append(1/x)
    x = x + 1


plt.scatter(x = x_list, y = y_list, color = "teal", edgecolor = "black")
plt.title("Graph of f(x) = 1/x")
plt.xlabel('x')
plt.ylabel("y")
plt.show()
````

````python
eps = 0.025

x_list = [] 
y_list = []


max_x = 100000

for x in range(1, max_x+1):
    if (1/x) <= eps:
        break
    x_list.append(x)
    y_list.append(1/x)

plt.scatter(x = x_list, y=y_list)
plt.xlabel("x values")
plt.ylabel("1/x")
plt.show()
````

### Success and Failure

````python
# The next example simulates a discrete-time Bernoulli process.
# For each time step, we will perform an experiment leading either 
# to "Success" or "Failure"

import random 

possible_outcomes = ["Success", "Failure"]

# experiment
outcome = random.choice(possible_outcomes)
print(outcome)
````

````python
# While Loop

success_count = 0
t = 0.0
final_time = 1.0
dt = 0.1

# while t != final_time:  # <--- WARNING: using this instead will cause infinite loop!

while abs(t - final_time) > 1e-13:
    # Experiment
    random_outcome = random.choice(possible_outcomes)

    # If it was a success, count it by increasing the success_count variable
    if random_outcome == "Success":
        success_count = success_count + 1

    # Increase time to the next value
    t = t + dt

print(success_count)
````

````python
def quadratic_formula(a, b, c):
    #body
    discrim = b**2 - 4*a*c
    x_plus = -b + np.sqrt(discrim)
    x_plus = x_plus/(2*a)

    x_minus = -b - np.sqrt(discrim)
    x_minus = x_minus/(2*a)

    #return two values as output
    return x_plus, x_minus

# Run
output = quadratic_formula(5,15,10) #Output is tuple
print(output[0])
print(output[1])

#or...
pos, neg = quadratic_formula(5,15,10)
print(pos)
````

### Function to return number of trials after 3 consecutive success

````python
# Solution 1:

def run_experiment(success_consec):
    num_success = 0
    num_trials = 0

    while num_success < 3:
        outcome = np.random.choice(["Success", "Failure"], p = [0.8, 0.2])
        num_trials = num_trials + 1
        print(outcome)

        if outcome == "Success":
            num_success = num_success + 1 
        else:
            num_success = 0
    return f"Number of trials it takes to get {success_consec} consecutive successes: {num_trials}"

print(run_experiment(3))
````

````python
# Solution 2:

def run_experiment(success_consec):
    num_success = []
    num_trials = 0

    while len(num_success) < 3:
        outcome = np.random.choice(["Success", "Failure"], p = [0.8, 0.2])
        num_trials = num_trials + 1
        print(outcome)

        if outcome == "Success":
            num_success.append("Success")
        else:
            num_success = []
    return f"Number of trials it takes to get {success_consec} consecutive successes: {num_trials}"

print(run_experiment(3))
````

````python

def num_til_consecutive(success_consec):
    num_success = 0
    num_trials = 0
    previous1 = False
    previous2 = False
   
    while num_success < 3:
        outcome = np.random.choice(["Success", "Failure"], p = [0.8, 0.2])
        num_trials = num_trials + 1
        print(outcome)

        if outcome == "Success" and previous1 == True and previous2 == True:
            break
        elif outcome == "Success" and previous1 == True:
            previous2 == True
            num_success = num_success + 1
        elif outcome == "Success":
            previous1 = True
            num_success = num_success + 1
        else: 
            num_success = 0
            previous1 = False
            previous2 = False

    return "Number of trials until " + str(success_consec) + " consecutive Successes: " + str(num_trials)

print(num_til_consecutive(3))
````

- Instead of using keywords like "mean", "min", "max", etc. we can use our own user-defined functions.

- In the next example, we will be aggregating over rows. This means we apply each aggregation function to a column of the DataFrame.

- So to count the # of unique driver Id's, we can write a function that takes in a DataFrame column (a Pandas Series) and returns the number of unique values

````python
def count_unique(col):
    return len(col.unique())

#what this does in agg:
count_unique(df_results["driverId"])

results_agg = df_results.agg(mean_points = ('points','mean'),
                          mean_laps =   ('laps','mean'),
                          min_points =  ('points','min'),
                          max_points =  ('points','max'),
                          num_drivers = ('driverId', count_unique))

display(results_agg)
````
